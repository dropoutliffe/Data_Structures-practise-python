```python
#gfg problem
arr = [8, 2, 4, 5, 3, 7, 1]

sol = Solution()
print(sol.missingNum(arr))

class Solution:
    def missingNum(self, arr):
        n = len(arr)+1

        expected_sum = (n* (n+1))//2
        actual_sum = sum(arr)

        return expected_sum - actual_sum

```
