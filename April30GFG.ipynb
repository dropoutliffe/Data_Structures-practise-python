```python
#gfg problem  array missing numbers
arr = [8, 2, 4, 5, 3, 7, 1]

sol = Solution()
print(sol.missingNum(arr))

class Solution:
    def missingNum(self, arr):
        n = len(arr)+1

        expected_sum = (n* (n+1))//2
        actual_sum = sum(arr)

        return expected_sum - actual_sum

```

```python
#bitwise approch
arr = [8, 2, 4, 5, 3, 7, 1]

sol = Solution()
print(sol.missingNum(arr))

class Solution:
    def missingNum(self, arr):
        n = len(arr)+1

        xor_arr = 0
        xor_all = 0


        for i in range(1,n+1):
            xor_all ^=i

        for num in arr:
            xor_arr ^=num


        return xor_all ^ xor_arr


```
