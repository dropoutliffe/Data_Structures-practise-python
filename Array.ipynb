```python
array = [4,5,2,4,7,23,22,1]

max_val == array[0]
for num in array:
  if num > max_val:
    max_val = num

print("maximum", max_val)
```

```python
#reverse the array

arr = [2,5,3,4,6,234,534,6354]

left , right = 0, len(arr)-1

while left < right:
  arr[left], arr[right] = arr[right], arr[left]
  left += 1
  right -= 1

print(arr)

```

```python
#second largest number in the array
def secon_largest(a, n):
  a.sort()

  seconlargest = -1
  for i in range(n-2,-1,-1):
   if a[i] != a[n-1]:
     seconlargest = a[i]
     break
  return seconlargest


if __name__ == '__main__':
  a = [34,57, 54, 4,44,89]

  n = len(a)
  answer = secon_largest(a, n)
  print("the answer", answer)
```
