```python
class Solution:
    def minJumps(self, arr):
        n = len(arr)

        # If first element is 0, we can't move
        if arr[0] == 0:
            return -1

        # If only one element
        if n == 1:
            return 0

        maxReach = arr[0]
        steps = arr[0]
        jumps = 1

        for i in range(1, n):

            # If we reached last index
            if i == n - 1:
                return jumps

            # Update maxReach
            maxReach = max(maxReach, i + arr[i])

            # Use a step
            steps -= 1

            # If no steps left → jump needed
            if steps == 0:
                jumps += 1

                # If current index is beyond maxReach → impossible
                if i >= maxReach:
                    return -1

                # Reset steps
                steps = maxReach - i

        return -1
```
